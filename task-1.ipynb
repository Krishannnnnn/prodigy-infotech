```python
import pandas as pd

df = pd.read_csv('/kaggle/input/datasets/vamsinossam/titanic-dataset-csv/Titanic-Dataset.csv')

print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Age'].dropna(), bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
df['Sex'].value_counts().plot(kind='bar')

plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```
